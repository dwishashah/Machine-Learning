ALGORITHM


1.   Find the entropy for data-set
2.   For every attribute/feature:

      a.calculate entropy for all categorical values 

      b.take avg information entropy for current attribute

      c.calculate gain for the current attribute
3.   Pick the highest gain attribute.
4.   Repeat until we get the decision tree.



```python
import pandas as pd
import numpy as np
import math
eps = np.finfo(float).eps
from numpy import log2 as log

data = pd.read_csv('/content/play_tennis.csv')
data.drop('day',axis=1,inplace=True)
data.head()
```

```python
def entropy1(data):
    attr = data.keys()[-1]   
    entropy = 0
    values = data[attr].unique()
    for value in values:
        fraction = data[attr].value_counts()[value]/len(data[attr])
        entropy += -fraction*np.log2(fraction)
    return entropy
  
  
def att_entropy(data,attribute):
  attr = data.keys()[-1]  
  target_variables = data[attr].unique()  
  variables = data[attribute].unique()    
  entropy2 = 0
  for variable in variables:
      entropy = 0
      for target_variable in target_variables:
          num = len(data[attribute][data[attribute]==variable][data[attr] ==target_variable])
          den = len(data[attribute][data[attribute]==variable])
          fraction = num/(den+eps)
          entropy += -fraction*log(fraction+eps)
      fraction2 = den/len(data)
      entropy2 += -fraction2*entropy
  return abs(entropy2)


def max_attr(data):
    Entropy_att = []
    IG = []
    for key in data.keys()[:-1]:

        IG.append(entropy1(data)-att_entropy(data,key))
    return data.keys()[:-1][np.argmax(IG)]
  
  
def attr_table(data, node,value):
  return data[data[node] == value].reset_index(drop=True)


def buildTree(data,tree=None): 
    attr = data.keys()[-1]   
    node = max_attr(data)
   
    attValue = np.unique(data[node])
    
    if tree is None:                    
        tree={}
        tree[node] = {}
    

    for value in attValue:
        
        subtable = attr_table(data,node,value)
        clValue,counts = np.unique(subtable['play'],return_counts=True)                        
        
        if len(counts)==1:
            tree[node][value] = clValue[0]                                                    
        else:        
            tree[node][value] = buildTree(subtable) 
                   
    return tree
```
